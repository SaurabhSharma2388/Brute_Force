# Brute Force Technique

Brute force algorithms are a straightforward approach to solving problems by trying all possible solutions until the correct one is found.

```python
import hashlib
import itertools
import string
```

```python
def hash_password(password):
    return hashlib.sha256(password.encode()).hexdigest()

def brute_force(target_hash, max_length):
    characters = string.ascii_lowercase + string.digits + string.ascii_uppercase
    attempts = 0
    for length in range(1, max_length + 1):
        for guess in itertools.product(characters, repeat=length):
            guess_password = ''.join(guess)
            guess_hash = hash_password(guess_password)
            attempts += 1
            print(f"Attempt #{attempts}: Trying Password '{guess_password}'")
            if guess_hash == target_hash:
                return guess_password, attempts
    return None, attempts
```

```python
hashed_password = hash_password('Ib3')  # Target hash for 'Ib3'
max_length = 3  # Maximum length of the password to try

cracked_password, attempts = brute_force(hashed_password, max_length)
if cracked_password:
    print(f"Password cracked in {attempts} attempts: {cracked_password}")
else:
    print(f"Password not cracked.")
```
